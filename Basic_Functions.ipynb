```python
import os
import pandas as pd

path = './Data/'

filename_read = os.path.join(path, "mpg.csv")
print(f"filename:{filename_read}")
df = pd.read_csv(filename_read)

print(df[0:5])
```

```python
# Basic statistics on the framework

df = pd.read_csv(filename_read, na_values = ['NA','?'])

# strip non-numeric

df = df.select_dtypes(include=['int','float'])
print(df[0:4])


headers = list(df.columns.values)
fields = []

for field in headers:
    fields.append({
        'name':field,
        'mean':df[field].mean(),
        'var':df[field].var(),
        'sdev':df[field].std()
    })
for field in fields:
    print(field)
```

```python
# Sort and shuffle

import numpy as np

np.random.seed(92)
df = df.reindex(np.random.permutation(df.index))
df.reset_index(inplace = True, drop = True)
df
```

```python
import os
import numpy as np
import pandas as pd

path = './Data/'

filename_read = os.path.join(path, "mpg.csv")
filename_save = os.path.join(path,"mpg_out.csv")

df = pd.read_csv(filename_read, na_values=['NA','?'])
df = df.reindex(np.random.permutation(df.index))

df.to_csv(filename_save, index = False) # not to write row numbers

df.drop('name',1, inplace = True)
print(f"After drop: {df.columns}")

# column inserting
df[0:5]

df.insert(1,"weight_kg", (df['weight']*0.4535).astype(int))
df


from scipy.stats import zscore

df['mpg'] = zscore(df['mpg'])
df[0:5]
```

```python
# missing values

med = df['horsepower']
df['horsepower'] = df['horsepower'].fillna(med)
pd.isnull(df['horsepower']).values.any()


col_horse = df['horsepower']
col_nm = df['origin']
result = pd.concat([col_horse,col_nm], axis=1)
result
```
